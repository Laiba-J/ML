```python
#replacing with NAN values
train.replace('?', np.nan)
```

```python
#checking null values
train.isnull().sum()
```

```python
#converting string into numeric values
train['normalized-losses'] = pd.to_numeric(train['normalized-losses'], errors='coerce')
```

```python
#performing mutation
train['normalized-losses'].mean()
train['normalized-losses'].fillna(train['normalized-losses'].mean())
```

```python
train.isnull().sum()
```

```python
#dropping null values
train.dropna(subset=['price'], inplace=True)
```
