```python
# checking if a number is positive, negative or zera
num=float(input("Enter a number: "))
if num>0:
  print(str(num)+" is positive")
elif num<0:
  print(str(num)+" is negative")
else:
  print(str(num)+" is zero")
```

```python
# interest calculator
princ=float(input("Enter the amount: "))
rate=float(input("Enter the rate of interest: "))
time=float(input("Enter the time(in years): "))
simple_int=(princ*rate*time)/100
print("The interest on this amount is: "+str(simple_int))
```
